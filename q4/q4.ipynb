```python
import csv
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib import font_manager, rc
font_path = "C:/Windows/Fonts/NGULIM.TTF"
font = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font)

f = open('202303_subway.csv', 'r', encoding='ANSI')
data = csv.reader(f)
next(data)
next(data)
index=[]
columns = ['riding', 'quiting', 'riding&quiting']
riding=[]
quiting=[]
rAndq=[]

for row in data:
    riding.append(int(row[4])+int(row[6]))
    quiting.append(int(row[5])+int(row[7]))
    rAndq.append(riding[-1]+quiting[-1])
    index.append(row[1]+" - "+row[3])
    
values = [riding, quiting, rAndq]
df = pd.DataFrame({'riding':riding, 'quiting':quiting,'riding&quiting':rAndq }, index=index)

plt.title('    30 ')
df = df.sort_values(by=['riding'], ascending=False)
tmp = df[:31]
plt.bar(tmp.index, tmp['riding'],0.25, label='riding')
plt.xticks(size=8,rotation=90)
plt.show()

plt.title('    30 ')
df = df.sort_values(by=['quiting'], ascending=False)
tmp = df[:31]
plt.bar(tmp.index, tmp['quiting'],0.25, label='quiting')
plt.xticks(size=8,rotation=90)
plt.show()

plt.title('    30 ')
df = df.sort_values(by=['riding&quiting'], ascending=False)
tmp = df[:31]
plt.bar(tmp.index, tmp['riding&quiting'],0.25, label='riding&quiting')
plt.xticks(size=8,rotation=90)
plt.show()

f.close()
```
