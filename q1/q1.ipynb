```python
import csv
import matplotlib.pyplot as plt

f_Korea = open('Korea.csv', 'r', encoding='ANSI')
data_Korea = csv.reader(f_Korea)
f_Seoul = open('Seoul.csv', 'r', encoding='ANSI')
data_Seoul = csv.reader(f_Seoul)
f_DaeJeon = open('DaeJeon.csv', 'r', encoding='ANSI')
data_DaeJeon = csv.reader(f_DaeJeon)
f_Busan = open('Busan.csv', 'r', encoding='ANSI')
data_Busan = csv.reader(f_Busan)
f_Jeju = open('Jeju.csv', 'r', encoding='ANSI')
data_Jeju = csv.reader(f_Jeju)

next(data_Korea)
next(data_Seoul)
next(data_DaeJeon)
next(data_Busan)
next(data_Jeju)
Korea = []
Seoul = []
DaeJeon = []
Busan = []
Jeju = []

for row in data_Korea:
    Korea.append(float(row[2]))
for row in data_Seoul:
    Seoul.append(float(row[2]))
for row in data_DaeJeon:
    DaeJeon.append(float(row[2]))
for row in data_Busan:
    Busan.append(float(row[2]))
for row in data_Jeju:
    Jeju.append(float(row[2]))
    
plt.title('2022 Korea Monthly Average Temperature')
plt.plot(range(1,13), Korea, label='Korea', color='red')
plt.plot(range(1,13), Seoul, label='Seoul', color='blue')
plt.plot(range(1,13), DaeJeon, label='DaeJeon', color='green')
plt.plot(range(1,13), Busan, label='Busan', color='orange')
plt.plot(range(1,13), Jeju, label='Jeju', color='purple')
plt.xlabel('month')
plt.ylabel('temperature')
plt.legend()
plt.show()
```

```python
diff = [0,0,0,0]
for i in range(12):
    diff[0] += (Korea[i] - Seoul[i])**2
    diff[1] += (Korea[i] - DaeJeon[i])**2
    diff[2] += (Korea[i] - Busan[i])**2
    diff[3] += (Korea[i] - Jeju[i])**2
    
print("전국 기준 서울 분산: {0:.1f}".format(diff[0]/12))
print("전국 기준 대전 분산: {0:.1f}".format(diff[1]/12))
print("전국 기준 부산 분산: {0:.1f}".format(diff[2]/12))
print("전국 기준 제주 분산: {0:.1f}".format(diff[3]/12))
```
