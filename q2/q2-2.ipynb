```python
import csv
import matplotlib.pyplot as plt

f= open('q2.csv', 'r', encoding='ANSI')
data = csv.reader(f)

cnt = 1
for row in data:
    title = "Dice Simulation {} tries".format((10**cnt)*10)
    cnt += 1
    plt.title(title)
    plt.hist(list(map(int,row[1:])), bins=6)
    plt.show()
    
f.close()
```
