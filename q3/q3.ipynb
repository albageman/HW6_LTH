```python
import csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

f = open('2018_2022_Jeju_population.csv', 'r', encoding='ANSI')
data = csv.reader(f)

next(data)
row = next(data)[1:]
man=[]
woman=[]
for i in range(5):
    row[i*4+2] = int(row[i*4+2].replace(",",""))
    man.append(row[i*4+2])
    row[i*4+3] = int(row[i*4+3].replace(",",""))
    woman.append(row[i*4+3])
    
year = ['2018','2019','2020','2021','2022']
df = pd.DataFrame({'man':man, 'woman':woman}, index=year)
fig, ax = plt.subplots(figsize=(12,6))
bar_width = 0.25
index = np.arange(5)

plt.bar(index, df['man'], bar_width, alpha=0.4, color='red', label='man')
plt.bar(index+bar_width, df['woman'], bar_width, alpha=0.4, color='blue', label='woman')
plt.xticks(np.arange(bar_width, 5+bar_width, 1),year)
plt.xlabel('year', size=13)
plt.ylabel('population', size=13)
plt.legend(loc='center right')
plt.show()

f.close()
```

```python
pop=[]
for i in range(5):
    pop.append(man[i]-woman[i])
    
plt.plot(year ,pop)
plt.xlabel('year')
plt.ylabel('man - woman')
plt.show()
```
